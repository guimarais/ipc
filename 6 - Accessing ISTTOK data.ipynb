# 6 - Accessing ISTTOK data

```python
from StartSdas import StartSdas
from LoadSdasData import LoadSdasData
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
client = StartSdas()
```

```python
shotnr = client.searchMaxEventNumber('0x0000')
shotnr = 36515
print(shotnr)
```

```python
plasma_curr_channelID='POST.PROCESSED.IPLASMA'; # Unique Identifier for plasma current
[ip,ipt] = LoadSdasData(client, plasma_curr_channelID, shotnr);
```

```python
plt.plot(ipt, ip)
plt.xlabel('time [ms]')
plt.ylabel('Ip [A]')
plt.show()
```

```python
shotnr = 37600
ne_channelID='POST.PROCESSED.DENSITY'; # Unique Identifier for plasma current
[ne,net] = LoadSdasData(client, ne_channelID, shotnr);
plt.plot(net, ne)
plt.xlabel('time [ms]')
plt.ylabel(r'$ne [m^{-3}]$')
plt.xlim(0,3e5)
plt.show()
```

```python
ne1_channelID = 'MARTE_NODE_IVO3.DataCollection.Channel_044';
ne2_channelID = 'MARTE_NODE_IVO3.DataCollection.Channel_045';
[ne1,net1] = LoadSdasData(client, ne1_channelID, shotnr);
[ne2,net2] = LoadSdasData(client, ne2_channelID, shotnr);
plt.plot(net1, ne1)
plt.plot(net2, ne2)
plt.xlabel('time [ms]')
plt.ylabel('ne []')
plt.ylim(0,2)
plt.show()
```

```python
vlp_channelID = 'PCIE_ATCA_ADC_16.BOARD_1.CHANNEL_026'
ilp_channelID = 'PCIE_ATCA_ADC_16.BOARD_1.CHANNEL_027'
[vlp, vlpt] = LoadSdasData(client, vlp_channelID, shotnr);
[ilp, ilpt] = LoadSdasData(client, ilp_channelID, shotnr);
plt.plot(ilpt, ilp)
plt.plot(vlpt, vlp)
plt.xlabel('time [ms]')
plt.ylabel('ne []')
plt.show()
```
